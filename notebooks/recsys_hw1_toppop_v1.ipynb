# VK RecSys HW-1 â€” TopPop baseline

Simple Top Popular recommender system.
- Implicit feedback (views)
- Recommends top-10 most popular items to each user
- Submission generation logic


```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
import os
os.listdir('/content/drive/MyDrive')
train = pd.read_parquet('/content/drive/MyDrive/VK/train.parquet')
test = pd.read_parquet('/content/drive/MyDrive/VK/test.parquet')
```

```python
print(len(train))
print(train.columns)
train.head()
```

```python
top_items =  train['item_id'].value_counts().head(10).index.tolist()
test_user = test['user_id'].unique()

recs = []
for user_id in test_user:
  for item_id in top_items:
    recs.append({
        'user_id': user_id,
        'recs' : item_id
    })

submission = pd.DataFrame(recs)
submission.head(10)
submission.to_csv('submission', index=False)
```
